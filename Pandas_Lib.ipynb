```python
#Pandas Series
#A Pandas Series is like a column in a table.A Pandas Series is like a column in a table.
#Create a simple Pandas Series from a list:
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)

import pandas as pd 
ab = ['a', 'b', 'c']
var=pd.Series(ab)
print(var)
```

```python
#Labels
#If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.
#This label can be used to access a specified value.
import pandas as pd 
ab = ['a', 'b', 'c']
var=pd.Series(ab)
print(var[1])

#Create Labels
#With the index argument, you can name your own labels.
#Create your own labels:
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)

import pandas as pd 
ab = ['a', 'b', 'c']
var=pd.Series(ab, index = ["Java", "Python", "C"])
print(var)

#accessing item by refering label
import pandas as pd 
ab = ['aaa', 'bbb', 'ccc']
var=pd.Series(ab, index = ["Java", "Python", "C"])
print(var["Java"])
```

```python
#Key/Value Objects as Series
#You can also use a key/value object, like a dictionary, when creating a Series.
#Create a simple Pandas Series from a dictionary:
import pandas as pd
calories = {"day1": 420, "day2": 380, "day3": 390}#The keys of the dictionary become the labels.
myvar = pd.Series(calories)
print(myvar)

#Create a Series using only data from "day1" and "day2":
import pandas as pd
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
```

```python
#DataFrames
#Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
#Series is like a column, a DataFrame is the whole table.
#create a DataFrame from two Series:
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
myvar = pd.DataFrame(data)
print(myvar)

import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
myvar = pd.DataFrame(data)
print(myvar)

import pandas as pd
dataframe = {
    "roll" : [12, 13, 14, 15],
    "name" : ["Akash", "Jay","Monali", "Raj" ]
}
var =pd.DataFrame(dataframe)
print(var)

#Locate Row
#Pandas use the loc attribute to return one or more specified row(s)
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
#load data into a DataFrame object:
df = pd.DataFrame(data)
print(df.loc[0])#returns a Pandas Series

import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
#load data into a DataFrame object:
df = pd.DataFrame(data)
print(df.loc[[0, 1]])#When using [], the result is a Pandas DataFrame.
```

```python
#Named Indexes
#With the index argument, you can name your own indexes.
#Add a list of names to give each row a name:

import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df) 

#Locate Named Indexes
#Use the named index in the loc attribute to return the specified row(s).
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df.loc["day2"])
```

```python
#Pandas - Cleaning Empty Cells
#Remove Rows
#One way to deal with empty cells is to remove rows that contain empty cells.
#Return a new Data Frame with no empty cells:
import pandas as pd
df = pd.read_csv('data.csv')
new_df = df.dropna()#the dropna() method returns a new DataFrame, and will not change the original.
print(new_df.to_string())
```

```python
#Remove all rows with NULL values:

import pandas as pd
df = pd.read_csv('data.csv')
df.dropna(inplace = True)#If you want to change the original DataFrame, use the inplace = True argument:
print(df.to_string())
```

```python
#Replace Only For Specified Columns
#Replace NULL values in the "Calories" columns with the number 130:

import pandas as pd

df = pd.read_csv('data1.csv')

df["NAME"].fillna(130, inplace = True)
print(df)
```

```python
#Replace Using Mean, Median, or Mode
#A common way to replace empty cells, is to calculate the mean, median or mode value of the column.
#Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column:
#Calculate the MEAN, and replace any empty values with it:

import pandas as pd
df = pd.read_csv('data1.csv')
x = df["ID"].mean()
df["ID"].fillna(x, inplace = True)
print(df)
```

```python
#Calculate the MEDIAN, and replace any empty values with it:
import pandas as pd
df = pd.read_csv('data1.csv')
x = df["ID"].median()
df["ID"].fillna(x, inplace = True)
print(df)
```

```python
#Calculate the MODE, and replace any empty values with it:
import pandas as pd
df = pd.read_csv('data1.csv')
x = df["ID"].mode()[0]
df["ID"].fillna(x, inplace = True)
print(df)
```
