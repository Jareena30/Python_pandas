```python
#Create an Empty Series
import pandas as pd
s = pd.Series()
print (s)
```

```python
#Create a Series from ndarray
import pandas as pd
import numpy as np
data = np.array(['a','b','c','d'])
data1 = np.array([2,3,'b',22,42])
s = pd.Series(data)
s1= pd.Series(data1)
print(s)
print(s1)
```

```python
#Create a Series from ndarray, index is passed
import pandas as pd
import numpy as np
data = np.array(['a','b','c','d'])
s = pd.Series(data,index=[100,101,102,103])
print(s)
```

```python
#Create a Series from ndarray
import pandas as pd
import numpy as np
data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data)
print(s)
```

```python
#Create a Series from ndarray, index is passed
import pandas as pd
import numpy as np
data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data,index=['b','c','d','a'])
print(s)
```

```python
#Create a Series from Scalar
import pandas as pd
import numpy as np
s = pd.Series(5, index=[0, 1, 2, 3])
print(s)
```

```python
#Accessing Data from Series with Position
#Retrieve the first element.
import pandas as pd
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

#retrieve the first element
print(s[0])
```

```python
#Accessing Data from Series with Position
#Retrieve the first three elements in the Series
import pandas as pd
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

#retrieve the first three element
print(s[:3])
```

```python
#Accessing Data from Series with Position
#Retrieve the last three elements in the Series
import pandas as pd
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

#retrieve the last three element
print(s[-3:])
```

```python
#Retrieve Data Using Label (Index)
#Retrieve a single element using index label value
import pandas as pd
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

#retrieve a single element
print(s['a'])
```

```python
#Retrieve Data Using Label (Index)
#Retrieve a multiple element using index label value
import pandas as pd
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

#retrieve multiple elements
print(s[['a','c','d']])
```

```python
#If a label is not contained, an exception is raised.

import pandas as pd
s = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])

#retrieve multiple elements
print(s['f'])
```
